```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/lena.png')

color_matrix=np.array(img)
print(color_matrix.shape)
plt.imshow(color_matrix)
```

```python
row=color_matrix.shape[0]
col=color_matrix.shape[1]

gray_matrix=np.zeros((row,col),dtype="int")
for i in range(row):
  for j in range(col):
    gray_matrix[i][j]=int(0.2989*color_matrix[i][j][0]+0.5870*color_matrix[i][j][1]+0.1140*color_matrix[i][j][2])

plt.imshow(gray_matrix,cmap="gray")
plt.show()
```

```python
def layer_image(bit_number,gray_matrix):
  print("Bit Layer:",bit_number)
  bit_matrix = np.zeros((row, col), dtype="int")
  for i in range(row):
    for j in range(col):
      bit_matrix[i][j] = gray_matrix[i][j] >> (bit_number-1)
      bit_matrix[i][j] = bit_matrix[i][j] & 1

  plt.imshow(bit_matrix, cmap="gray")
  plt.show()
```

```python
for bit_number in range(1,9):
  layer_image(bit_number,gray_matrix)
```
