```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/lena.png')

color_matrix=np.array(img)
print(color_matrix.shape)
plt.imshow(color_matrix)
```

```python
black_white=np.zeros_like(color_matrix)
row=color_matrix.shape[0]
col=color_matrix.shape[1]
for i in range(row):
  for j in range(col):
    if color_matrix[i][j][1]>127:
      black_white[i][j]=[255,255,255]
    else:
      black_white[i][j]=[0,0,0]


plt.imshow(black_white,cmap="gray")
plt.show()
```

```python
row=color_matrix.shape[0]
col=color_matrix.shape[1]

gray_matrix=np.zeros((row,col),dtype="int")
for i in range(row):
  for j in range(col):
    gray_matrix[i][j]=int(0.2989*color_matrix[i][j][0]+0.5870*color_matrix[i][j][1]+0.1140*color_matrix[i][j][2])

plt.imshow(gray_matrix,cmap="gray")
plt.show()
```

```python
row=gray_matrix.shape[0]
col=gray_matrix.shape[1]

binary_matrix=np.zeros((row,col),dtype="int")
for i in range(row):
  for j in range(col):
    if gray_matrix[i][j]>127:
      binary_matrix[i][j]=255
    else:
      binary_matrix[i][j]=0

plt.imshow(binary_matrix,cmap="gray")
plt.show()
```
